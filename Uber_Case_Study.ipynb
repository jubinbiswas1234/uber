```python
import numpy as np 
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ub= pd.read_csv(r"C:\Users\koust\Downloads\Uber Case Study\Uber Request Data.csv")
ub.head()
```

```python

ub['Request timestamp']=pd.to_datetime(ub['Request timestamp'])
ub['Drop timestamp']=pd.to_datetime(ub['Drop timestamp'])
ub.head()
```

```python
ub.info()
```

```python
ub.isna().sum()
```

```python
ub['Request Date']=ub['Request timestamp'].dt.date
ub['Drop Date']=ub['Drop timestamp'].dt.date
ub['Request Time']=ub['Request timestamp'].dt.time
ub['Drop Time']=ub['Drop timestamp'].dt.time
ub['Request Day']=ub['Request timestamp'].dt.day
ub['Drop Day']=ub['Drop timestamp'].dt.day
ub['Request Hour']=ub['Request timestamp'].dt.hour
ub.head(50)
```

```python
ub['Status'].value_counts()
```

```python
ub['Status'].value_counts().plot.bar().label=True
```

```python
ub["Time Slot"]= pd.cut(ub['Request Hour'], bins=5,labels= False) 
ub.groupby(['Time Slot'])
ub.head()
```

```python
x = ub.groupby(['Time Slot', 'Request Hour','Status']).sum()
x
```

```python
ub.groupby(['Time Slot']).count()
```

```python
ub['Time Slot'].value_counts().plot.bar().label=True
plt.title('Most Requests')
```

```python
ub['Request Hour'].value_counts().plot.bar().label=True
plt.title('Hourly Requests')
```

```python
grp=ub.groupby(["Time Slot","Request Hour"]).Status.count()
sns.barplot(ub['Status'],ub['Time Slot'])
plt.title('TimeSlotwise Status')
plt.show()
```
